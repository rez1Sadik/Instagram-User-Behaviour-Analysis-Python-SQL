```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

#comments
def read_data_from_csv():
    comments=pd.read_csv('comments.csv')
    return comments


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    comments=read_data_from_csv()

    #Remove Unwanted columns
    comments.columns = comments.columns.str.strip()
    comments.drop(columns=["posted date","emoji used", "Hashtags used count"],axis=1,inplace=True)
    
    #rename columns
    comments.rename(columns={'comment':"comment_text",'User  id':"user_id",'Photo id':"photo_id", 'created Timestamp':"created_at"},inplace=True)
   
    #export cleaned Dataset to newcsv file named "comments_cleaned.csv"
    comments.to_csv('comments_cleaned.csv')
    return comments


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()

#follows
def read_data_from_csv():
    follows=pd.read_csv('follows.csv')
    return follows


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    follows=read_data_from_csv()

    #Remove Unwanted columns
    follows.columns = follows.columns.str.strip()
    follows.drop(columns=['is follower active', 'followee Acc status'],axis=1,inplace=True)
    
    #rename columns
    follows.rename(columns={'follower':"follower_id",'followee':"followee_id",'created time':"created_at"},inplace=True)
    
    #export cleaned Dataset to newcsv file named "follows_cleaned.csv"
    follows.to_csv('follows_cleaned.csv')
    return follows


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()


#likes
def read_data_from_csv():
    likes=pd.read_csv('likes.csv')
    return likes


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    likes=read_data_from_csv()

    #Remove Unwanted columns
    likes.columns = likes.columns.str.strip()
    likes.drop(columns=['following or not', 'like type'],axis=1,inplace=True)
    
    #rename columns
    likes.rename(columns={'user':"user_id",'photo':"photo_id",'created time':"created_at"},inplace=True)
    
    #export cleaned Dataset to newcsv file named "likes_cleaned.csv"
    likes.to_csv('likes_cleaned.csv')
    return likes


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()


#photo_tags

def read_data_from_csv():
    photo_tags=pd.read_csv('photo_tags.csv')
    return photo_tags


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    photo_tags=read_data_from_csv()

    #Remove Unwanted columns
    photo_tags.columns = photo_tags.columns.str.strip()
    photo_tags.drop(columns=[ 'user id'],axis=1,inplace=True)
    
    #rename columns
    photo_tags.rename(columns={'tag ID':"tag_id",'photo':"photo_id"},inplace=True)    
    
    #export cleaned Dataset to newcsv file named "photo_tags_cleaned.csv"
    photo_tags.to_csv('photo_tags_cleaned.csv')
    return photo_tags


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()



#photos
def read_data_from_csv():
    photos=pd.read_csv('photos.csv')
    return photos


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    photos=read_data_from_csv()

    #Remove Unwanted columns
    photos.columns = photos.columns.str.strip()
    photos.drop(columns=['Insta filter used','photo type'],axis=1,inplace=True)
    
     #rename columns
    photos.rename(columns={'user ID':"user_id",'image link':"image_url",'created dat':"created_date"},inplace=True)

    
    #export cleaned Dataset to newcsv file named "photos_cleaned.csv"
    photos.to_csv('photos_cleaned.csv')
    return photos


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()

#tags
def read_data_from_csv():
    tags=pd.read_csv('tags.csv')
    return tags


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    tags=read_data_from_csv()

    #Remove Unwanted columns
    tags.columns = tags.columns.str.strip()
    tags.drop(columns=["location"],axis=1,inplace=True)
    
    #rename columns
    tags.rename(columns={'tag text':"tag_name",'created time':"created_at"},inplace=True)
    
    #export cleaned Dataset to newcsv file named "tags_cleaned.csv"
    tags.to_csv('tags_cleaned.csv')
    return tags


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()

#users

def read_data_from_csv():
    users=pd.read_csv('users.csv')
    return users


def data_cleaning():
    #call remove_unwanted_columns() function to get dataframe
    users=read_data_from_csv()

    #Remove Unwanted columns
    users.columns = users.columns.str.strip()
    users.drop(columns=['private/public', 'post count','Verified status'],axis=1,inplace=True)
    
    #rename columns
    users.rename(columns={'name':"username",'created time':"created_at"},inplace=True)
  

    #export cleaned Dataset to newcsv file named "users_cleaned.csv"
    users.to_csv('users_cleaned.csv')
    return users


#Do not Delete the Following function
def task_runner():
    data_cleaning()

task_runner()
```
